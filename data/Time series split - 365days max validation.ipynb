```python
from sklearn.model_selection import TimeSeriesSplit, train_test_split
import numpy as np

total_folds = 5
timeSeriesCrossValidation = TimeSeriesSplit(n_splits=total_folds)
```

```python
data_length = 365 * (2018 - 1990)
x_data = np.random.rand(data_length, 51 + 43)
y_data = np.random.rand(data_length)

print("Fold size: ", x_data.shape[0] / (total_folds + 1))
```

```python
def reduce_validation_fold_size(train, validation, max_validation_length=365 ):
    if len(validation) > max_validation_length:
        train = np.append(train, validation[0:len(validation)- max_validation_length])
        validation = validation[-max_validation_length:]
    return train, validation
```

```python
k = timeSeriesCrossValidation.split(x_data, y_data)
k.first
```

```python
counter = 0
max_validation_length = 365
for train, validation in timeSeriesCrossValidation.split(x_data, y_data):  # TODO: test train/dev/validation
    counter += 1
    print("\n\nFold: ", counter)
    
    train, validation = reduce_validation_fold_size(train, validation)
    

    print("Train ({}):".format(len(train)), train,"\nValidation ({}): ".format(len(validation)), validation)
```
