## LS on Rosenbrock

```python
import random
import numpy as np
from scipy.optimize import minimize
from scipy.optimize import Bounds
```

```python
import scipy
scipy.__version__
```

```python
def benchmark_function(x):
    a = 0.6
    b = 33
    if x[0] > 1.1:
        print("Outside bounds x[0]", x[0])
    if x[1] > 1.1:
        print("Outside bounds x[1]", x[1])
    if x[2] > 1.1:
        print("Outside bounds x[2]", x[2])
    return sum(b * (x[1:]-x[:-1]**2.0)**2.0 + (a - x[:-1])**2.0) # Rosenbrock
    #return sum(x**2.0) # Sphere
    #return (x[:-1] + 2 * x[1:] - 7) **2.0 + (2 * x[:-1] + x[1:] - 5)**2 # Booth
```

```python
data_manipulation = {}
data_manipulation["rank"] = 1
iterations = 1000

#bounds = [[-10, 10], [-10, 10]]
#x0 = [random.uniform(0, 1), random.uniform(0, 1)]
bounds = [[-3, 1], [-10, 1], [-10, 1]]
scipy_bounds = Bounds([bounds[0][0], bounds[1][0], bounds[2][0]], [bounds[0][1], bounds[1][1], bounds[2][1]])

x0 = [random.uniform(bounds[0][0], bounds[0][1]), random.uniform(bounds[1][0], bounds[1][1]), random.uniform(bounds[2][0], bounds[2][1])]

# Bounded local optimizers
local_search_method = "L-BFGS-B"

for i in range(1, 9):
    data_manipulation["rank"] = i
    # Bounded local optimizers
    local_search_method = "L-BFGS-B"
    options = dict(maxfun=iterations)
    if data_manipulation["rank"] % 10 == 1:
        local_search_method = "L-BFGS-B"
        options = dict(maxfun=iterations)
    elif data_manipulation["rank"] % 10 == 2:
        local_search_method = "SLSQP"
        options = dict(maxiter=iterations)
    elif data_manipulation["rank"] % 10 == 3:
        local_search_method = "TNC"
        options = dict(maxiter=iterations)
    elif data_manipulation["rank"] % 10 == 4:
        local_search_method = "trust-constr"
        options = dict(maxiter=iterations)          
    elif data_manipulation["rank"] % 10 == 5:
        local_search_method = "Nelder-Mead"  
        options = dict(maxiter=iterations)          
    elif data_manipulation["rank"] % 10 == 6:
        local_search_method = "Powell"    
        options = dict(maxiter=iterations)
    elif data_manipulation["rank"] % 10 == 7:
        local_search_method = "CG"  
        options = dict(maxiter=iterations)  
    elif data_manipulation["rank"] % 10 == 8:
        local_search_method = "COBYLA"
        options = dict(maxiter=iterations)

    print("\n=== {}\n".format(local_search_method))

    res = minimize(
        x0=x0,
        fun=benchmark_function,
        method=local_search_method,
        bounds=scipy_bounds,
        options=options
    )

    print(res)
```
